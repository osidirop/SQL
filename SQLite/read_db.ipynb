```python
import pandas as pd
import sqlite3 as sql
```

```python
# establish the connection with the db
connection = sql.connect('chinook.db')

# SQL statements to get all the 11 tables
tb_media_types = 'select * from media_types'
tb_employees = 'select * from employees'
tb_customers = 'select * from customers'
tb_invoices = 'select * from invoices'
tb_invoice_items = 'select * from invoice_items'
tb_artists = 'select * from artists'
tb_albums = 'select * from albums'
tb_genres = 'select * from genres'
tb_tracks = 'select * from tracks'
tb_playlists = 'select * from playlists'
tb_playlist_track = 'select * from playlist_track'

# Get tables into pandas dataframes
df_media_types = pd.read_sql(tb_media_types, connection)
df_employees = pd.read_sql(tb_employees, connection)
df_customers = pd.read_sql(tb_customers, connection)
df_invoices = pd.read_sql(tb_invoices, connection)
df_invoice_items = pd.read_sql(tb_invoice_items, connection)
df_artists = pd.read_sql(tb_artists, connection)
df_albums = pd.read_sql(tb_albums, connection)
df_genres = pd.read_sql(tb_genres, connection)
df_tracks = pd.read_sql(tb_tracks, connection)
df_playlists = pd.read_sql(tb_playlists, connection)
df_playlist_track = pd.read_sql(tb_playlist_track, connection)

# close the connection
connection.close()

# print a df
df_employees.head()
```
